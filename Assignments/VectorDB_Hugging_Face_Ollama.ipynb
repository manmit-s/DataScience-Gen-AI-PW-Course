# Question 1: What is a Vector Database (VectorDB) and how is it different from traditional databases?

## What is a Vector Database?
A **Vector Database** stores and retrieves **vector embeddings** (high-dimensional numerical representations of data like text, images, or audio produced by ML models).  
It enables **similarity search**, finding items that are semantically similar rather than exactly identical.

Similarity is usually measured using:
- Cosine similarity
- Euclidean distance
- Dot product

## Difference from Traditional Databases

| Feature | Traditional Database | Vector Database |
|-------|---------------------|----------------|
| Data Type | Structured rows/columns | High-dimensional vectors |
| Query Type | Exact match (SQL) | Similarity search (ANN) |
| Search Basis | Keywords / values | Semantic similarity |
| Index Type | B-tree, Hash | HNSW, IVF, LSH |
| Use Case | CRUD, transactions | AI search, recommendations |

**Example:**  
Searching *"heart attack"* in a traditional DB returns exact matches.  
A VectorDB can also return *"myocardial infarction"* because the embeddings are semantically similar.

# Question 3: Why is ChromaDB important in AI/ML projects?

**ChromaDB** is a lightweight, open-source vector database widely used in **LLM and RAG applications**.

## Key Advantages
- Simple setup (`pip install chromadb`)
- Native integration with **LangChain and LlamaIndex**
- Supports **in-memory and persistent storage**
- Built-in embedding generation
- Metadata filtering with vector search
- Collections for organizing vectors
- Supports cosine, L2, and inner product similarity

**Why developers use it:**  
It allows rapid prototyping of AI applications without managing complex vector infrastructure.

````python
# Q6
# Install: pip install chromadb sentence-transformers
!pip install chromadb
import chromadb
from chromadb.utils import embedding_functions

# 1. Initialize Chroma client (persistent storage)
client = chromadb.PersistentClient(path="./chroma_store")

# 2. Use Sentence Transformer for embeddings
embedding_fn = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)

# 3. Create a collection
collection = client.get_or_create_collection(
    name="medical_articles",
    embedding_function=embedding_fn,
    metadata={"hnsw:space": "cosine"}
)

# 4. Sample documents
documents = [
    "Heart attack occurs when blood flow to the heart is blocked.",
    "Diabetes is a disease that affects blood sugar levels.",
    "Hypertension is a condition where blood pressure is consistently high.",
    "Asthma causes airways to become inflamed and narrow.",
    "Cancer is characterized by uncontrolled cell growth.",
]

metadatas = [
    {"topic": "cardiology"},
    {"topic": "endocrinology"},
    {"topic": "cardiology"},
    {"topic": "pulmonology"},
    {"topic": "oncology"},
]

ids = [f"doc_{i}" for i in range(len(documents))]

# 5. Insert documents
collection.add(
    documents=documents,
    metadatas=metadatas,
    ids=ids
)

print(f"Total documents in collection: {collection.count()}")

# 6. Semantic Search
query = "myocardial infarction and cardiac issues"
results = collection.query(
    query_texts=[query],
    n_results=3
)

print("\n--- Semantic Search Results ---")
print(f"Query: {query}\n")
for i, (doc, meta, dist) in enumerate(zip(
    results["documents"][0],
    results["metadatas"][0],
    results["distances"][0]
)):
    print(f"Result {i+1}:")
    print(f"  Document : {doc}")
    print(f"  Topic    : {meta['topic']}")
    print(f"  Distance : {dist:.4f}")
    print()

# 7. Metadata-filtered search
print("--- Filtered Search (cardiology only) ---")
filtered = collection.query(
    query_texts=["blood pressure and heart"],
    n_results=2,
    where={"topic": "cardiology"}
)
for doc in filtered["documents"][0]:
    print(f"  > {doc}")
````

````python
# Q7
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling
)
from datasets import Dataset
import torch

# 1. Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
print(f"Loading model: {model_name}")

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token  # GPT2 has no pad token by default

model = AutoModelForCausalLM.from_pretrained(model_name)
print("Model loaded successfully!")
print(f"Model parameters: {model.num_parameters():,}")

# 2. Custom training data (medical Q&A style)
train_texts = [
    "Patient presents with chest pain. Diagnosis: The patient may be experiencing a myocardial infarction.",
    "Symptoms include high blood sugar and frequent urination. Diagnosis: These are classic signs of diabetes mellitus.",
    "Patient reports shortness of breath during exercise. Diagnosis: This could indicate early-stage heart failure.",
    "Persistent cough with wheezing observed. Diagnosis: Patient likely has bronchial asthma.",
    "Blood pressure reading of 160/100 mmHg. Diagnosis: Stage 2 hypertension confirmed.",
    "Patient complains of extreme fatigue and weight gain. Diagnosis: Possible hypothyroidism should be investigated.",
]

# 3. Tokenize the dataset
def tokenize(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        max_length=128,
        padding="max_length"
    )

dataset = Dataset.from_dict({"text": train_texts})
tokenized_dataset = dataset.map(tokenize, batched=True, remove_columns=["text"])
tokenized_dataset = tokenized_dataset.add_column("labels", tokenized_dataset["input_ids"])

print(f"\nDataset size: {len(tokenized_dataset)} samples")

# 4. Training Arguments
training_args = TrainingArguments(
    output_dir="./gpt2-medical-finetuned",
    num_train_epochs=3,
    per_device_train_batch_size=2,
    save_steps=10,
    logging_steps=5,
    learning_rate=5e-5,
    warmup_steps=2,
    prediction_loss_only=True,
    report_to="none"  # Disable wandb
)

# 5. Data Collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # Causal LM, not masked LM
)

# 6. Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
)

print("\nStarting fine-tuning...")
trainer.train()
print("Fine-tuning complete!")

# 7. Save the fine-tuned model
model.save_pretrained("./gpt2-medical-finetuned")
tokenizer.save_pretrained("./gpt2-medical-finetuned")
print("Model saved to ./gpt2-medical-finetuned")

# 8. Generate text with fine-tuned model
print("\n--- Text Generation with Fine-Tuned Model ---")
model.eval()

prompts = [
    "Patient presents with chest pain.",
    "Symptoms include high blood sugar",
]

for prompt in prompts:
    inputs = tokenizer(prompt, return_tensors="pt")
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=40,
            do_sample=True,
            temperature=0.7,
            pad_token_id=tokenizer.eos_token_id
        )
    generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nPrompt   : {prompt}")
    print(f"Generated: {generated}")
````

````python
# Q9
# Install: pip install chromadb sentence-transformers ollama
# Terminal: ollama pull llama3

import ollama
import chromadb
from chromadb.utils import embedding_functions

# ─────────────────────────────────────────────
# STEP 1: Set up ChromaDB as the Vector Store
# ─────────────────────────────────────────────
print("Initializing ChromaDB...")
client = chromadb.PersistentClient(path="./rag_chroma_store")

embedding_fn = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)

collection = client.get_or_create_collection(
    name="knowledge_base",
    embedding_function=embedding_fn,
    metadata={"hnsw:space": "cosine"}
)

# ─────────────────────────────────────────────
# STEP 2: Load Knowledge Base (Medical Articles)
# ─────────────────────────────────────────────
knowledge_base = [
    {
        "id": "art_001",
        "text": "Myocardial infarction, commonly known as a heart attack, occurs when coronary arteries become blocked by plaque buildup. Symptoms include chest pain, shortness of breath, and pain radiating to the left arm. Immediate treatment involves aspirin, nitroglycerin, and emergency angioplasty.",
        "source": "Cardiology Journal 2023"
    },
    {
        "id": "art_002",
        "text": "Type 2 diabetes is characterized by insulin resistance. The pancreas initially compensates by producing more insulin but eventually cannot maintain normal blood glucose levels. Lifestyle modifications including diet and exercise are first-line treatments. Metformin is the most commonly prescribed medication.",
        "source": "Endocrinology Review 2023"
    },
    {
        "id": "art_003",
        "text": "Hypertension affects one in three adults globally. Blood pressure above 130/80 mmHg is classified as hypertension. ACE inhibitors, calcium channel blockers, and diuretics are common first-line medications. Lifestyle changes like reducing sodium intake and regular exercise significantly lower blood pressure.",
        "source": "Internal Medicine Bulletin 2023"
    },
    {
        "id": "art_004",
        "text": "Asthma is a chronic inflammatory disease of the airways. Triggers include allergens, exercise, cold air, and stress. Management involves short-acting bronchodilators (rescue inhalers) for acute episodes and inhaled corticosteroids for long-term control. Severe asthma may require biologic therapies.",
        "source": "Pulmonology Today 2023"
    },
    {
        "id": "art_005",
        "text": "COVID-19 is caused by the SARS-CoV-2 virus. Common symptoms include fever, cough, fatigue, and loss of taste or smell. Severe cases may lead to pneumonia and acute respiratory distress syndrome (ARDS). Vaccines have been proven effective in reducing severe disease and hospitalization rates.",
        "source": "Infectious Disease Report 2023"
    },
]

# Insert into ChromaDB if not already present
if collection.count() == 0:
    collection.add(
        documents=[doc["text"] for doc in knowledge_base],
        metadatas=[{"source": doc["source"]} for doc in knowledge_base],
        ids=[doc["id"] for doc in knowledge_base]
    )
    print(f"Inserted {len(knowledge_base)} documents into knowledge base.")
else:
    print(f"Knowledge base already loaded: {collection.count()} documents.")

# ─────────────────────────────────────────────
# STEP 3: Define RAG Pipeline Functions
# ─────────────────────────────────────────────
def retrieve_context(query: str, n_results: int = 2) -> list[dict]:
    """Retrieve top-k relevant documents from ChromaDB."""
    results = collection.query(
        query_texts=[query],
        n_results=n_results
    )

    retrieved = []
    for doc, meta, dist in zip(
        results["documents"][0],
        results["metadatas"][0],
        results["distances"][0]
    ):
        retrieved.append({
            "text": doc,
            "source": meta["source"],
            "similarity": 1 - dist  # Convert distance to similarity
        })

    return retrieved


def build_prompt(query: str, context_docs: list[dict]) -> str:
    """Build RAG prompt by combining query with retrieved context."""
    context_str = ""
    for i, doc in enumerate(context_docs):
        context_str += f"\n[Context {i+1} - Source: {doc['source']}]\n{doc['text']}\n"

    prompt = f"""You are a helpful medical assistant. Use ONLY the provided context to answer the question.
If the context doesn't contain enough information, say so clearly.

CONTEXT:
{context_str}

QUESTION: {query}

ANSWER (based on the context above):"""

    return prompt


def rag_query(query: str) -> dict:
    """Full RAG pipeline: retrieve → augment → generate."""
    print(f"\n{'='*60}")
    print(f"QUERY: {query}")
    print('='*60)

    # Step 1: Retrieve
    context_docs = retrieve_context(query, n_results=2)
    print(f"\n[RETRIEVAL] Found {len(context_docs)} relevant documents:")
    for i, doc in enumerate(context_docs):
        print(f"  [{i+1}] {doc['source']} (similarity: {doc['similarity']:.3f})")
        print(f"       {doc['text'][:80]}...")

    # Step 2: Augment (Build Prompt)
    prompt = build_prompt(query, context_docs)

    # Step 3: Generate with Llama3 via Ollama
    print("\n[GENERATION] Llama3 generating response...")
    response = ollama.chat(
        model="llama3",
        messages=[{"role": "user", "content": prompt}],
        options={"temperature": 0.3}  # Lower temp for factual accuracy
    )

    answer = response["message"]["content"]

    return {
        "query": query,
        "retrieved_docs": context_docs,
        "answer": answer
    }

# ─────────────────────────────────────────────
# STEP 4: Run RAG Queries
# ─────────────────────────────────────────────
queries = [
    "What are the treatment options for heart attack?",
    "How is type 2 diabetes managed?",
    "What medications are used for hypertension?",
]

for query in queries:
    result = rag_query(query)
    print(f"\n[ANSWER]:\n{result['answer']}")
    print("-" * 60)
```

**Expected Output:**
```
Initializing ChromaDB...
Inserted 5 documents into knowledge base.

============================================================
QUERY: What are the treatment options for heart attack?
============================================================

[RETRIEVAL] Found 2 relevant documents:
  [1] Cardiology Journal 2023 (similarity: 0.891)
       Myocardial infarction, commonly known as a heart attack, occurs when coronary...
  [2] Internal Medicine Bulletin 2023 (similarity: 0.523)
       Hypertension affects one in three adults globally. Blood pressure above...

[GENERATION] Llama3 generating response...

[ANSWER]:
Based on the provided context, treatment options for a heart attack include:
1. Immediate aspirin administration to prevent further clotting
2. Nitroglycerin to relieve chest pain and improve blood flow
3. Emergency angioplasty to reopen the blocked coronary artery
Early intervention is critical for minimizing heart muscle damage.
------------------------------------------------------------
````

````python
# Q10
# Install: pip install chromadb sentence-transformers transformers ollama torch
# Terminal: ollama pull llama3

import ollama
import chromadb
from chromadb.utils import embedding_functions
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore")

print("=" * 65)
print("   HEALTHBOT: Medical Research Article Q&A System")
print("=" * 65)

# ─────────────────────────────────────────────────────────────
# COMPONENT 1: ChromaDB (VectorDB for Retrieval)
# ─────────────────────────────────────────────────────────────
print("\n[1/3] Setting up ChromaDB Vector Store...")

client = chromadb.PersistentClient(path="./healthbot_store")
embedding_fn = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)
collection = client.get_or_create_collection(
    name="medical_research",
    embedding_function=embedding_fn,
    metadata={"hnsw:space": "cosine"}
)

# Medical Research Articles Knowledge Base
medical_articles = [
    {
        "id": "research_001",
        "title": "Cardiovascular Disease Prevention",
        "content": """Cardiovascular disease (CVD) remains the leading cause of death worldwide.
        Primary prevention involves lifestyle modifications: regular aerobic exercise (150 minutes/week),
        Mediterranean diet rich in omega-3 fatty acids, smoking cessation, and maintaining BMI under 25.
        Statins are recommended for patients with LDL above 190 mg/dL. Aspirin therapy is indicated
        for high-risk patients. Regular blood pressure monitoring and keeping systolic below 130 mmHg
        significantly reduces CVD risk by 25%.""",
        "journal": "NEJM 2023"
    },
    {
        "id": "research_002",
        "title": "Diabetes Management Guidelines",
        "content": """Type 2 diabetes management focuses on glycemic control with HbA1c target below 7%.
        Metformin remains first-line therapy due to its safety profile and cardiovascular benefits.
        SGLT-2 inhibitors (empagliflozin, dapagliflozin) and GLP-1 receptor agonists (semaglutide)
        have shown cardioprotective and renoprotective effects in clinical trials. Continuous glucose
        monitoring (CGM) improves glycemic control. Structured diabetes education programs reduce
        hospitalizations by 30%.""",
        "journal": "Diabetes Care 2023"
    },
    {
        "id": "research_003",
        "title": "COVID-19 Long-term Effects",
        "content": """Post-acute sequelae of SARS-CoV-2 (PASC), commonly known as Long COVID,
        affects approximately 10-30% of COVID-19 survivors. Persistent symptoms include fatigue,
        brain fog, dyspnea, and post-exertional malaise lasting more than 12 weeks. Mechanisms
        include viral persistence, immune dysregulation, and microbiome disruption. Current
        management is symptomatic. Multidisciplinary Long COVID clinics provide comprehensive care.
        Paxlovid (nirmatrelvir-ritonavir) reduces hospitalization risk by 89% in high-risk groups.""",
        "journal": "Lancet 2023"
    },
    {
        "id": "research_004",
        "title": "Mental Health and Exercise",
        "content": """Regular physical exercise demonstrates significant antidepressant and anxiolytic effects.
        Meta-analyses show exercise reduces depression symptoms as effectively as antidepressants in
        mild-to-moderate cases. Mechanisms include increased BDNF (brain-derived neurotrophic factor),
        endorphin release, and HPA axis regulation. Recommended: 30 minutes of moderate-intensity
        aerobic exercise 3-5 times per week. Yoga and mindfulness-based interventions reduce cortisol
        levels by 20% and improve sleep quality significantly.""",
        "journal": "JAMA Psychiatry 2023"
    },
    {
        "id": "research_005",
        "title": "Antibiotic Resistance Crisis",
        "content": """Antimicrobial resistance (AMR) poses a critical global health threat. WHO estimates
        700,000 deaths annually due to drug-resistant infections, projected to reach 10 million by 2050.
        MRSA, multi-drug resistant tuberculosis, and carbapenem-resistant Enterobacteriaceae are
        major concerns. Antibiotic stewardship programs reduce inappropriate use by 30-50%. New
        antibiotics in development include cefiderocol and imipenem-cilastatin-relebactam. Phage
        therapy shows promise for resistant infections.""",
        "journal": "WHO Bulletin 2023"
    },
    {
        "id": "research_006",
        "title": "Cancer Immunotherapy Advances",
        "content": """Immune checkpoint inhibitors (ICIs) have revolutionized cancer treatment.
        PD-1/PD-L1 inhibitors (pembrolizumab, nivolumab) show durable responses in melanoma,
        NSCLC, and bladder cancer. CAR-T cell therapy achieves complete remission in 50-90% of
        relapsed/refractory B-cell lymphomas. Combination therapies with chemotherapy and targeted
        agents improve outcomes. Biomarkers including TMB (tumor mutational burden) and MSI-H
        status predict ICI response. Immune-related adverse events require early recognition and
        steroid management.""",
        "journal": "Cancer Cell 2023"
    },
]

if collection.count() == 0:
    collection.add(
        documents=[a["content"] for a in medical_articles],
        metadatas=[{"title": a["title"], "journal": a["journal"]} for a in medical_articles],
        ids=[a["id"] for a in medical_articles]
    )
    print(f"  Loaded {len(medical_articles)} medical research articles into ChromaDB.")
else:
    print(f"  ChromaDB ready: {collection.count()} articles loaded.")

# ─────────────────────────────────────────────────────────────
# COMPONENT 2: Hugging Face — Query Understanding & Classification
# ─────────────────────────────────────────────────────────────
print("\n[2/3] Loading Hugging Face Zero-Shot Classifier...")

# HuggingFace used for: understanding query intent & medical topic classification
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"  # Lightweight zero-shot model
)

MEDICAL_TOPICS = [
    "cardiovascular disease",
    "diabetes",
    "COVID-19",
    "mental health",
    "antibiotic resistance",
    "cancer treatment",
    "general medicine"
]

def classify_query(query: str) -> dict:
    """Use HuggingFace to classify the medical topic of the query."""
    result = classifier(query, MEDICAL_TOPICS, multi_label=False)
    return {
        "top_topic": result["labels"][0],
        "confidence": result["scores"][0],
        "all_scores": dict(zip(result["labels"], result["scores"]))
    }

print("  HuggingFace classifier loaded successfully.")

# ─────────────────────────────────────────────────────────────
# COMPONENT 3: Ollama Llama3 — Answer Generation
# ─────────────────────────────────────────────────────────────
print("\n[3/3] Connecting to Ollama (Llama3)...")
try:
    test = ollama.chat(model="llama3", messages=[{"role":"user","content":"ping"}])
    print("  Ollama Llama3 connected successfully.")
except:
    print("  Warning: Ollama not running. Ensure: ollama serve && ollama pull llama3")

# ─────────────────────────────────────────────────────────────
# CORE RAG PIPELINE
# ─────────────────────────────────────────────────────────────
def retrieve_articles(query: str, n_results: int = 2) -> list[dict]:
    """Retrieve relevant medical articles from ChromaDB."""
    results = collection.query(
        query_texts=[query],
        n_results=n_results
    )
    retrieved = []
    for doc, meta, dist in zip(
        results["documents"][0],
        results["metadatas"][0],
        results["distances"][0]
    ):
        retrieved.append({
            "content": doc,
            "title": meta["title"],
            "journal": meta["journal"],
            "relevance": round((1 - dist) * 100, 1)
        })
    return retrieved


def generate_answer(query: str, context_docs: list[dict], topic_info: dict) -> str:
    """Generate medically accurate answer using Llama3 + retrieved context."""
    context_str = ""
    for i, doc in enumerate(context_docs):
        context_str += f"""
Research Article {i+1}: "{doc['title']}"
Published in: {doc['journal']}
Content: {doc['content']}
"""

    system_prompt = """You are HealthBot, an AI assistant for a health-tech startup.
You answer medical questions based STRICTLY on the provided research articles.
Rules:
- Only use information from the provided research articles
- Always cite the journal source
- Be concise but comprehensive
- Always end with: "⚠️ Consult a licensed physician for personalized medical advice."
- Never make up information not in the articles"""

    user_prompt = f"""Research Articles:
{context_str}

User Question: {query}
Detected Medical Topic: {topic_info['top_topic']} (confidence: {topic_info['confidence']:.0%})

Please answer the question based on the research articles above."""

    response = ollama.chat(
        model="llama3",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        options={"temperature": 0.2, "num_predict": 300}
    )

    return response["message"]["content"]


def healthbot_query(user_query: str) -> None:
    """Complete HealthBot pipeline for a user query."""
    print(f"\n{'━'*65}")
    print(f"👤 USER: {user_query}")
    print('━'*65)

    # Step 1: HuggingFace Query Classification
    topic_info = classify_query(user_query)
    print(f"\n🔍 [HuggingFace Analysis]")
    print(f"   Detected Topic : {topic_info['top_topic']}")
    print(f"   Confidence     : {topic_info['confidence']:.1%}")

    # Step 2: ChromaDB Retrieval
    articles = retrieve_articles(user_query, n_results=2)
    print(f"\n📚 [ChromaDB Retrieval] - {len(articles)} relevant articles found:")
    for a in articles:
        print(f"   • {a['title']} ({a['journal']}) — Relevance: {a['relevance']}%")

    # Step 3: Llama3 Generation
    print(f"\n🤖 [Ollama Llama3 Generating Answer...]")
    answer = generate_answer(user_query, articles, topic_info)

    print(f"\n💬 HEALTHBOT RESPONSE:")
    print(f"{'─'*65}")
    print(answer)
    print('━'*65)

# ─────────────────────────────────────────────────────────────
# DEMO: Run Sample Medical Queries
# ─────────────────────────────────────────────────────────────
print("\n" + "=" * 65)
print("   HEALTHBOT DEMO - Sample Medical Queries")
print("=" * 65)

sample_queries = [
    "What lifestyle changes can prevent heart disease?",
    "What are the best medications for type 2 diabetes in 2023?",
    "What is Long COVID and how is it treated?",
]

for query in sample_queries:
    healthbot_query(query)
    print()

# Interactive mode
print("\n" + "=" * 65)
print("   INTERACTIVE MODE (type 'exit' to quit)")
print("=" * 65)
while True:
    user_input = input("\n👤 Your medical question: ").strip()
    if user_input.lower() in ["exit", "quit", "q"]:
        print("Thank you for using HealthBot. Stay healthy! 🏥")
        break
    if user_input:
        healthbot_query(user_input)
```

**Expected Output:**
```
═══════════════════════════════════════════════════════════════
   HEALTHBOT: Medical Research Article Q&A System
═══════════════════════════════════════════════════════════════

[1/3] Setting up ChromaDB Vector Store...
  Loaded 6 medical research articles into ChromaDB.

[2/3] Loading Hugging Face Zero-Shot Classifier...
  HuggingFace classifier loaded successfully.

[3/3] Connecting to Ollama (Llama3)...
  Ollama Llama3 connected successfully.

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
👤 USER: What lifestyle changes can prevent heart disease?
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

🔍 [HuggingFace Analysis]
   Detected Topic : cardiovascular disease
   Confidence     : 94.2%

📚 [ChromaDB Retrieval] - 2 relevant articles found:
   • Cardiovascular Disease Prevention (NEJM 2023) — Relevance: 89.3%
   • Mental Health and Exercise (JAMA Psychiatry 2023) — Relevance: 61.2%

🤖 [Ollama Llama3 Generating Answer...]

💬 HEALTHBOT RESPONSE:
─────────────────────────────────────────────────────────────
According to research published in NEJM 2023, the following lifestyle
changes significantly reduce cardiovascular disease risk:

1. **Regular Exercise**: 150 minutes/week of aerobic activity
2. **Diet**: Mediterranean diet rich in omega-3 fatty acids
3. **Smoking Cessation**: One of the most impactful changes
4. **Weight Management**: Maintaining BMI under 25
5. **Blood Pressure Control**: Keep systolic below 130 mmHg

These measures can reduce CVD risk by up to 25%.

⚠️ Consult a licensed physician for personalized medical advice.
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
```

---

**Architecture Summary for Q10:**
```
User Query
    │
    ▼
[Hugging Face - bart-large-mnli]
    │  Zero-shot classification
    │  → Detects medical topic + confidence
    │
    ▼
[ChromaDB - VectorDB]
    │  Semantic similarity search
    │  → Retrieves top-2 relevant research articles
    │
    ▼
[Ollama - Llama3]
    │  Context-aware generation
    │  → Produces grounded, cited answer
    │
    ▼
Final Answer (cited, grounded, safe)
````

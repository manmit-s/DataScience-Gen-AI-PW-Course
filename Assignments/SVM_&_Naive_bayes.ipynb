## Q1. What is a Support Vector Machine (SVM)
- A supervised learning algorithm used for classification and regression.
- It finds the optimal hyperplane that separates classes with the maximum margin.

## Q2. Difference Between Hard Margin and Soft Margin SVM
- **Hard Margin**: Assumes perfect linear separability; no misclassification allowed.
- **Soft Margin**: Allows some misclassification; introduces a penalty term to handle noisy data.

## Q3. Mathematical Intuition Behind SVM
- Maximizes the margin between support vectors and the decision boundary.
- Solves a convex optimization problem using quadratic programming.

## Q4. Role of Lagrange Multipliers in SVM
- Converts constrained optimization into an unconstrained dual problem.
- Helps derive the support vectors and decision boundary.

## Q5. What are Support Vectors in SVM
- Data points closest to the decision boundary.
- They directly influence the position and orientation of the hyperplane.

## Q6. What is a Support Vector Classifier (SVC)
- SVM variant used for classification tasks.
- Can handle both linear and non-linear classification using kernels.

##  Q7. What is a Support Vector Regressor (SVR)
- SVM variant used for regression tasks.
- Fits the best line within a threshold margin (epsilon).

## Q8. What is the Kernel Trick in SVM
- Transforms data into higher dimensions to make it linearly separable.
- Uses kernel functions to avoid explicit computation of transformation.

## Q9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel

| Kernel Type     | Description                              | Use Case                          |
|----------------|------------------------------------------|-----------------------------------|
| Linear          | No transformation; fast and simple       | Linearly separable data           |
| Polynomial      | Maps to polynomial space                 | Complex but structured data       |
| RBF (Gaussian)  | Infinite-dimensional mapping             | Non-linear and complex boundaries |

## Q10. Effect of the C Parameter in SVM
- Controls trade-off between margin size and classification error.
- **Low C**: Wider margin, more tolerance to misclassification.
- **High C**: Narrow margin, less tolerance to misclassification.

## Q11. Role of Gamma in RBF Kernel SVM
- Defines influence of a single training example.
- **Low Gamma**: Far-reaching influence, smoother decision boundary.
- **High Gamma**: Close influence, more complex boundary.

## Q12. What is the Naïve Bayes Classifier and Why is it “Naïve”
- A probabilistic classifier based on Bayes’ Theorem.
- Called “naïve” because it assumes feature independence.

##  Q13. Bayes’ Theorem



\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]



## Q14. Differences Between Gaussian, Multinomial, and Bernoulli Naïve Bayes

| Variant               | Description                                  | Use Case                          |
|----------------------|----------------------------------------------|-----------------------------------|
| Gaussian             | Assumes features follow normal distribution  | Continuous data                   |
| Multinomial          | Based on word counts                         | Text classification               |
| Bernoulli            | Binary features (0/1)                        | Spam detection, sentiment analysis|

## Q15. When to Use Gaussian Naïve Bayes
- When features are continuous and approximately normally distributed.

## Q16. Key Assumptions Made by Naïve Bayes
- Features are conditionally independent given the class.
- All features contribute equally and independently to the outcome.

##  Q17. Advantages and Disadvantages of Naïve Bayes

**Advantages**:
- Fast and scalable
- Works well with high-dimensional data
- Performs well with small datasets

**Disadvantages**:
- Assumes feature independence (often unrealistic)
- Struggles with correlated features

##  Q18. Why Naïve Bayes is Good for Text Classification
- Text data often has high dimensionality and sparse features.
- Multinomial Naïve Bayes handles word frequencies effectively.

## Q19. Compare SVM and Naïve Bayes for Classification Tasks

| Aspect              | SVM                                  | Naïve Bayes                          |
|---------------------|---------------------------------------|--------------------------------------|
| Type                | Discriminative                       | Generative                           |
| Performance         | High accuracy, slower training       | Fast training, good baseline         |
| Feature Assumption  | No independence assumption           | Assumes feature independence         |
| Use Case            | Complex boundaries, image/text       | Text classification, spam detection |

##  Q20. How Laplace Smoothing Helps in Naïve Bayes
- Prevents zero probability for unseen features.
- Adds a small constant (usually 1) to all counts to ensure non-zero probabilities.


```python
# Q21. Train an SVM Classifier on the Iris dataset and evaluate accuracy
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
clf = SVC()
clf.fit(X_train, y_train)
print("Accuracy:", accuracy_score(y_test, clf.predict(X_test)))
```

```python
# Q22. Train SVM with Linear and RBF kernels on Wine dataset and compare accuracy
from sklearn.datasets import load_wine
wine = load_wine()
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=42)
linear = SVC(kernel='linear').fit(X_train, y_train)
rbf = SVC(kernel='rbf').fit(X_train, y_train)
print("Linear Accuracy:", accuracy_score(y_test, linear.predict(X_test)))
print("RBF Accuracy:", accuracy_score(y_test, rbf.predict(X_test)))
```

```python
# Q23. Train SVR on housing dataset and evaluate using MSE
from sklearn.datasets import fetch_california_housing
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

housing = fetch_california_housing()
X_train, X_test, y_train, y_test = train_test_split(housing.data, housing.target, test_size=0.3, random_state=42)
svr = SVR()
svr.fit(X_train, y_train)
print("MSE:", mean_squared_error(y_test, svr.predict(X_test)))
```

```python
# Q24. Train SVM with Polynomial Kernel and visualize decision boundary
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1)
clf = SVC(kernel='poly', degree=3).fit(X, y)
plot_decision_regions(X, y, clf=clf)
plt.title("Polynomial Kernel SVM")
plt.show()
```

```python
# Q25. Train Gaussian Naïve Bayes on Breast Cancer dataset and evaluate accuracy
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)
gnb = GaussianNB().fit(X_train, y_train)
print("Accuracy:", accuracy_score(y_test, gnb.predict(X_test)))
```

```python
# Q26. Train Multinomial Naïve Bayes on 20 Newsgroups dataset
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

newsgroups = fetch_20newsgroups(subset='train')
X = CountVectorizer().fit_transform(newsgroups.data)
clf = MultinomialNB().fit(X, newsgroups.target)
print("Training Accuracy:", clf.score(X, newsgroups.target))
```

```python
# Q27. Train SVM with different C values and compare decision boundaries
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC

C_vals = [0.1, 1, 10]
for c in C_vals:
    X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=42)
    clf = SVC(C=c).fit(X, y)
    plot_decision_regions(X, y, clf=clf)
    plt.title(f"SVM with C={c}")
    plt.show()
```

```python
# Q35. Feature selection before Naïve Bayes
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)

X_new_train = SelectKBest(chi2, k=10).fit_transform(X_train, y_train)
X_new_test = SelectKBest(chi2, k=10).fit_transform(X_test, y_test)

clf = GaussianNB().fit(X_new_train, y_train)
print("Accuracy after feature selection:", accuracy_score(y_test, clf.predict(X_new_test)))
```

```python
# Q46. Visualize SVM Precision-Recall Curve
from sklearn.metrics import precision_recall_curve
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)
svc = SVC().fit(X_train, y_train)

precision, recall, _ = precision_recall_curve(y_test, svc.decision_function(X_test))
plt.plot(recall, precision, marker='.')
plt.title("SVM Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.grid(True)
plt.show()
```

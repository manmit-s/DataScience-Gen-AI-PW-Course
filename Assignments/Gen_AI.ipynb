## Q1: What are Large Language Models (LLMs) and how do they function?
- LLMs are AI models trained on massive text datasets to understand and generate human-like language.  
- They use the **Transformer architecture** with self-attention to capture relationships between words.  
- Training: Predicts the next token in text, learning grammar, meaning, and patterns.  
- Inference: Input → tokenization → embeddings → transformer layers → output generated step-by-step.  
- Key features: positional encoding, multi-head attention, probabilistic token prediction.

---

## Q2: Impact of LLMs on traditional software development
- Shift from explicit coding to **prompting** in natural language.  
- Faster prototyping with code generation (functions, APIs, tests).  
- Enables conversational interfaces instead of complex UIs.  
- Encourages modular backends where LLMs orchestrate tools and APIs.  
- Outputs are probabilistic → need strong testing, validation, and human oversight.

---

## Q3: Advantages & Limitations of LLMs
**Advantages**  
- Versatile across tasks (summarization, translation, Q&A).  
- Zero/few-shot learning without retraining.  
- Boosts productivity (drafting, documentation, support).  

**Limitations**  
- Can hallucinate incorrect info.  
- Inherit biases from training data.  
- Require heavy compute resources.  
- Privacy risks if sensitive data is shared.  
- Knowledge may be outdated without retrieval/fine-tuning.

---

## Q4: Industry Applications
- **Healthcare**: Documentation, triage assistants, billing automation.  
- **Finance**: Chatbots, fraud detection, risk analysis.  
- **E-commerce**: Personalized recommendations, conversational shopping.  
- **Education**: Tutoring systems, grading help, content summarization.

## Q5: LangChain vs LlamaIndex
- **LangChain**: General-purpose framework for building LLM apps.  
  - Focus: chaining prompts, tools, agents, workflows.  
  - Best for complex pipelines and multi-step reasoning.  

- **LlamaIndex**: Specialized in data indexing & retrieval.  
  - Focus: connecting LLMs to external data sources.  
  - Best for document Q&A, semantic search, knowledge bases.  

**Comparison**  
- LangChain → broad toolkit for orchestration.  
- LlamaIndex → efficient search & retrieval layer.

```python
#Q7
import requests
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Weather API call
def get_weather(city):
    url = f"http://api.weatherapi.com/v1/current.json?key=YOUR_API_KEY&q={city}"
    return requests.get(url).json()["current"]["condition"]["text"]

# LLM integration
llm = OpenAI(model="text-davinci-003")
prompt = PromptTemplate(
    input_variables=["city","weather"],
    template="The weather in {city} is {weather}. Write a short travel tip."
)
chain = LLMChain(llm=llm, prompt=prompt)

city = "Mumbai"
weather = get_weather(city)
print(chain.run({"city": city, "weather": weather}))
```

```python
#Q9
from llama_index import GPTVectorStoreIndex, SimpleDirectoryReader
from langchain.llms import OpenAI

# Load and index documents
documents = SimpleDirectoryReader("data").load_data()
index = GPTVectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

# Chatbot function
def chatbot(question):
    response = query_engine.query(question)
    return response

print(chatbot("What is the main topic of the file?"))
```

```python
#Q10
from llama_index import GPTVectorStoreIndex, SimpleDirectoryReader
from langchain.llms import OpenAI

# Load legal docs
documents = SimpleDirectoryReader("legal_docs").load_data()
index = GPTVectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

# Summarization
llm = OpenAI(model="text-davinci-003")
summary = llm("Summarize the key clauses in these documents.")
print(summary)

# Retrieval
response = query_engine.query("Find documents related to contract termination")
print(response)
```

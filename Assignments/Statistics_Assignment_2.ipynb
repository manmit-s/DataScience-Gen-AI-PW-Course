### Question 1: What is hypothesis testing in statistics?
**Answer:**  
Hypothesis testing is a statistical procedure to make inferences or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses—the null hypothesis and the alternative hypothesis—and then using sample evidence to decide whether to reject the null hypothesis.

---

### Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
**Answer:**  
The null hypothesis (denoted as \( H_0 \)) is a statement that there is no effect or no difference, and it represents the default or status quo position. The alternative hypothesis (\( H_1 \) or \( H_a \)) is a statement that there is an effect or a difference. The null hypothesis is tested directly, and statistical evidence is used to determine if it should be rejected in favor of the alternative hypothesis.

---

### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
**Answer:**  
The significance level, denoted by \( \alpha \), is the probability of rejecting the null hypothesis when it is actually true (Type I error). It acts as a threshold for decision-making in hypothesis testing—if the p-value is less than or equal to \( \alpha \), the null hypothesis is rejected; otherwise, it is not rejected.

---

### Question 4: What are Type I and Type II errors? Give examples of each.
**Answer:**  
- **Type I Error:** Rejecting the null hypothesis when it is actually true (a false positive).
  - *Example:* Concluding that a medicine is effective when it actually is not.
- **Type II Error:** Failing to reject the null hypothesis when the alternative hypothesis is true (a false negative).
  - *Example:* Concluding that a medicine is not effective when it actually is effective.


---

### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
**Answer:**  
A Z-test is used when the population variance is known and the sample size is large (\( n > 30 \)). A T-test is used when the population variance is unknown and/or the sample size is small (\( n \leq 30 \)). Use a Z-test for large samples with known variance and a T-test for small samples or when the population variance is unknown.

```python
# Question 7 Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.

import numpy as np
from scipy import stats

sample_scores = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9])

population_mean = 80

sample_mean = np.mean(sample_scores)
sample_std = np.std(sample_scores, ddof=1)


n = len(sample_scores)
z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

p_value = stats.norm.sf(z_statistic)

alpha = 0.05  # Significance level

print(f"Sample Mean: {np.mean(sample_scores)}")
print(f"Population Mean (under H0): {population_mean}")
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")
print(f"Significance level (alpha): {alpha}")

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is sufficient evidence to suggest the sample mean is greater than the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is not enough evidence to suggest the sample mean is greater than the population mean.")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mu = 50  # Population mean
sigma = 10  # Population standard deviation
sample_size = 100  # Number of samples
simulated_data = np.random.normal(mu, sigma, sample_size)


sample_mean = np.mean(simulated_data)
sample_std = np.std(simulated_data, ddof=1) # Use ddof=1 for sample standard deviation

standard_error = sample_std / np.sqrt(sample_size)

alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)
margin_of_error = t_critical * standard_error

confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Simulated Data Mean: {sample_mean}")
print(f"Simulated Data Standard Deviation: {sample_std}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f})")

plt.hist(simulated_data, bins=20, density=True, alpha=0.6, color='g')
plt.xlabel("Value")
plt.ylabel("Frequency (Density)")
plt.title("Histogram of Simulated Normal Distribution Data")
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def calculate_z_scores(data):
  mean = np.mean(data)
  std_dev = np.std(data)

  if std_dev == 0:
      return np.zeros_like(data)

  z_scores = (data - mean) / std_dev
  return z_scores

titanic = sns.load_dataset('titanic')

data_to_standardize = titanic['age'].dropna()

z_scores = calculate_z_scores(data_to_standardize)

plt.hist(z_scores, bins=20, density=True, alpha=0.6, color='purple')
plt.xlabel("Z-score")
plt.ylabel("Frequency (Density)")
plt.title("Histogram of Z-scores (Titanic Age Data)")
plt.show()

print("Explanation of Z-scores:")
print("A Z-score represents the number of standard deviations a particular data point is away from the mean of the dataset.")
print("- A Z-score of 0 means the data point is equal to the mean.")
print("- A positive Z-score indicates the data point is above the mean.")
print("- A negative Z-score indicates the data point is below the mean.")
print("For example, a Z-score of 1.5 means the data point is 1.5 standard deviations above the mean.")
```

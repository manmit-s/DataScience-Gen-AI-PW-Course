```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
subjects = ["Maths", "CS", "English", "Data Science", "Science"]
marks = [64.3, 92.8, 88, 90.5, 72]
academics_marks = pd.DataFrame(marks, columns = ["Marks"], index = subjects)
print(academics_marks.Marks.mean())
```

```python
# 3. Create a scatter plot using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x = x, y = y, color = "teal")
plt.title("Scatter Plot of Random Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips[["tip", "size", "total_bill"]]
df.corr()
sns.heatmap(df.corr(), cmap = "coolwarm", annot = True)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([1,2,3,4])
y1 = np.array([2,3,9,4])

x2 = np.array([1,2,3,4])
y2 = np.array([23,65,39,69])

plt.plot(x1 , y1, color = "maroon", label='Line 1', marker = "o")
plt.plot(x2, y2, color="teal", label='Line 2', marker = "o")
plt.legend()
plt.title("Line Plot with Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
subjects = ["Maths", "CS", "English", "Data Science", "Science"]
marks = [44.3, 22.8, 48, 70, 72] # marks out of 80
threshold_marks = 35
academics_marks = pd.DataFrame(marks, columns = ["Marks"], index = subjects)
academics_marks["pass_remark"] = academics_marks.Marks.apply(lambda x : "Pass" if x > threshold_marks else "Fail")
academics_marks
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.
l2 = np.random.randn(50)
df = pd.DataFrame(l2, columns=["Values"])

sns.histplot(data=df, x="Values", bins=15, color="orange")
plt.title("Histogram of Values (Normal Dist.)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
